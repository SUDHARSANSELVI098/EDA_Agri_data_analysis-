```python
#1
#Top 7 RICE PRODUCTION State Data(Bar_plot)

fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
grouped = df.groupby('state_name')['rice_production_(1000_tons)'].sum()
top7 = grouped.nlargest(7)
top7.plot(kind='bar',color='blue')
plt.title("Top 7 States by Rice Production")
plt.xlabel("STATES")
plt.ylabel("RICE PRODUCTION IN TONS")
plt.show()

```

```python
#1Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
#fig = plt.figure()
fig, ax = plt.subplots()

ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
grouped = df.groupby('state_name')['wheat_production_(1000_tons)'].sum()
top5 = grouped.nlargest(5)
top5.plot(kind='bar',color='green')
plt.title("Top 5 States by Rice Production")
plt.xlabel("STATES")
plt.ylabel("WHEAT PRODUCTION IN TONS)")
plt.show()
```

```python
# 2percentage of wheat production
fig = plt.figure()
fig.patch.set_facecolor('Teal')
ax = top5.plot(kind='pie', autopct='%1.1f%%')
ax.set_title("TOP 5 STATES BY WHEAT PRODUCTION PERCENTAGES ")
plt.ylabel('')  
plt.show()
```

```python
#3 Oil seed production by top 5 states
fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Teal')
oil_group=df.groupby('state_name')['oilseeds_production_(1000_tons)'].sum()
Large=oil_group.nlargest(5)
Large.plot(kind='bar',color='red')
plt.title('TOP 5 OILSEEDS PRODUCTION IN TONS')
plt.ylabel('OILSEEDS PRODUCTION IN TONS')
plt.xlabel('STATES')
```

```python
#4 Top 7 SUNFLOWER PRODUCTION  State
fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
Sunflower_top7=df.groupby('state_name')['sunflower_production_(1000_tons)'].sum()
Top=Sunflower_top7.nlargest(7)
Top.plot(kind='bar',color='purple')
plt.title("Top 7 SUNFLOWER PRODUCTION  STATES")
plt.ylabel('SUNFLOWER PRODUCTION IN TONS')
plt.xlabel("STATES")
```

```python
#5 India's SUGARCANE PRODUCTION From Last 50 Years
fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
plt.title('INDIA SUGARCANE PRODUCTION From Last 50 Years')
plt.ylabel('SUGARCANE PRODUCTION IN TONS')
sugar=df.groupby('year')['sugarcane_production_(1000_tons)'].sum()
sugar.plot(kind='line',color='darkred')
plt.xlabel('YEARS')
```

```python
#6 Rice Production Vs Wheat Production (Last 50y)
fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
plt.title('RICE PRODUCTION VS WHEAT PRODCUTION Last 50 Years')
group_rice=df.groupby('state_name')['rice_production_(1000_tons)'].sum()
group_wheat=df.groupby('state_name')['wheat_production_(1000_tons)'].sum()
group_rice.plot(kind='line',color='darkred')
group_wheat.plot(kind='line',color='darkblue')
plt.xlabel('STATES')
plt.ylabel('PRODUCTION IN TONS')
plt.legend()
plt.show()
```

```python
#7 Rice Production By West Bengal Districts
fig, ax = plt.subplots()
ax.set_facecolor('lightblue') 
fig.patch.set_facecolor('Gray')
state=df[df['state_name'] == 'West Bengal']
group_r=state.groupby('dist_name')['rice_production_(1000_tons)'].sum()
group_r.plot(kind='bar',color='purple')
plt.title('RICE PRODUCTION BY WEST BENGAL DISTRICTS')
plt.xlabel('WEST BENGAL DISTRICTS')
plt.ylabel('RICE PRODUCTION IN TONS')

```

```python
#8 Top 10 Wheat Production Years From UP
fig,ax=plt.subplots()
ax.set_facecolor('lightblue')
fig.patch.set_facecolor('grey')
state=df[df['state_name']=='Uttar Pradesh']
group_w=state.groupby('year')['wheat_production_(1000_tons)'].sum()
top10=group_w.nlargest(10)
top10.plot(kind='bar',color='yellow')
plt.title('TOP 10 WHEAT PRODUCTION YEARS FROM UP')
plt.xlabel('YEARS')
plt.ylabel('WHEAT PRODUCTION')
```

```python
#9 Millet Production (Last 50y)
fig,ax=plt.subplots()
ax.set_facecolor('lightblue')
fig.patch.set_facecolor('grey')
group=df.groupby('year')['finger_millet_production_(1000_tons)'].sum()
group1=df.groupby('year')['pearl_millet_production_(1000_tons)'].sum()
group.plot(kind='line',color='red')
group1.plot(kind='line',color='green')
plt.title("MILLETS PRODUCTIN ")
plt.xlabel('YEARS')
plt.ylabel("MILLETS")
plt.legend()
```

```python
#10 Top 7 States for Groundnut Production
fig,ax=plt.subplots()
ax.set_facecolor("skyblue")
fig.patch.set_facecolor("grey")
g=df.groupby('state_name')['groundnut_production_(1000_tons)'].sum()
r=g.nlargest(7)
r.plot(kind='bar',color='black')
plt.title("TOP 7 STATES FOR GROUNDNUT PRODUCTION")
plt.xlabel("STATES")
plt.ylabel("GROUNDNUT PRODUCTION IN TONS")
```

```python
#11
fig, ax1 = plt.subplots()
ax1.set_facecolor("skyblue")
fig.patch.set_facecolor("grey")
g = df.groupby('state_name')['soyabean_production_(1000_tons)'].sum()
r = g.nlargest(7)
r.plot(kind='bar', ax=ax1, color='orange')
ax1.set_xlabel("States")
ax1.set_ylabel("Soybean Production (1000 tons)", color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
ax2 = ax1.twinx()
yield_eff = df.groupby('state_name')['soyabean_yield_(kg_per_ha)'].mean()
yield_r = yield_eff.loc[r.index]
yield_r.plot(kind='line', ax=ax2, color='green', marker='o')
ax2.set_ylabel("Yield Efficiency (kg/ha)", color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title("Top 7 States: Soybean Production and Yield Efficiency")
plt.show()
```
